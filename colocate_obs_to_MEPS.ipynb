```python
import numpy as np
import netCDF4
import sys
#sys.path.append('/uio/kant/geo-metos-u7/franzihe/Documents/Thesis/Python')
sys.path.append('/home/franzihe/Documents/Python/Thesis')
import createFolder as cF
```

```python
year = '2016'
station = 'Haukeliseter'
```

```python
#Minutely obs
obs_f = dict()
obs_double_fence_acc = dict()
obs_wind_dir_10m   = dict()
obs_wind_speed_10m = dict()
obs_double_fence_temp   = dict()
obs_time = dict()
obs_dofe_acc = dict()
obs_wind_dir = dict()
obs_wind_speed = dict()
obs_air_temp = dict()
obs_time_ = dict()

#Eklima
eklima_f = dict()
eklima_double_fence_acc = dict()
eklima_wind_dir_10m = dict()
eklima_wind_speed_10m = dict()
eklima_air_temp = dict()
eklima_time = dict()
eklima_dofe_acc = dict()
eklima_wind_dir = dict()
eklima_wind_speed = dict()
eklima_air_temp = dict()
eklima_time_ = dict()

#Retrieval
ret_f = dict()
ret_double_fence_p = dict()
ret_double_fence_u = dict()
ret_dofe_acc_p = dict()
ret_dofe_acc_u = dict()

SWC_f = dict()
SWC_p = dict()
SWC_u = dict()
SWC_vert_p = dict()
SWC_vert_u = dict()

MRR_f = dict()
ZE = dict()
MRR_height = dict()
MRR_time = dict()
Ze_vert = dict()
MRR_h = dict()
```

```python
m = ['11', '12', '01', '02', '03']
```

```python
for month in m:
    if month == '11':
        t = np.arange(8,31)
    if month == '12' or month == '01' or month == '03':
        t = np.arange(1,32)
    if month == '02':
        t = np.arange(1,29)
    if month == '11' or month == '12':
        year = '2016'
    if month == '01' or month == '02' or month == '03':
        year = '2017'
    

    for day in t:
        
        if day < 10:
            day = '0%s' %day
        else:
            day = '%s' %day
            
        try:    
            # Minutely observation data
         #   obs_dirnc = '../../../Data/Weathermast/%s/' %station
            obs_dirnc = '/home/franzihe/Documents/Data/Weathermast/%s/' %station
            obs_f[int(year+month+day)] = netCDF4.Dataset(obs_dirnc+'Met_%s%s%s_%s.nc' %(year,month,day,station), 'r')

            obs_double_fence_acc[int(year+month+day)] = (obs_f[int(year+month+day)].variables['double_fence_accum'][:])   #Total accumulation (bucket value) from Geonor inside double fence [mm] RA1, RA2, RA3
            obs_wind_dir_10m[int(year+month+day)]   = (obs_f[int(year+month+day)].variables['10m_wind_dir_mast1'][:])                              #wind direction 10 m @mast 1 [deg] DD
            obs_wind_speed_10m[int(year+month+day)] = (obs_f[int(year+month+day)].variables['10m_wind_speed_mast1'][:])                            #wind speed 10 m @ mast 2 [m/s] X1FF
            obs_double_fence_temp[int(year+month+day)] = (obs_f[int(year+month+day)].variables['air_temp'][:])                                     #Air temperature, PT100 [deg C] TA
            obs_time[int(year+month+day)] = (obs_f[int(year+month+day)].variables['sec'][:])

          #  obs_f[int(year+month+day)].close()

            # Eklima data
            eklima_dirnc = '/home/franzihe/Documents/Data/Eklima/%s/' %station
            eklima_f[int(year+month+day)] = netCDF4.Dataset(eklima_dirnc+'precip_wind_temp_%s%s%s.nc' %( year,month,day), 'r')

            eklima_double_fence_acc[int(year+month+day)] = (eklima_f[int(year+month+day)].variables['RR_1'][:])
            eklima_wind_dir_10m[int(year+month+day)]   = (eklima_f[int(year+month+day)].variables['DD'][:])
            eklima_wind_speed_10m[int(year+month+day)] = (eklima_f[int(year+month+day)].variables['FF'][:])
            eklima_air_temp[int(year+month+day)] = (eklima_f[int(year+month+day)].variables['TA'][:])
            eklima_time[int(year+month+day)] = (eklima_f[int(year+month+day)].variables['time_stamp'][:])

         #   eklima_f[int(year+month+day)].close()


            # Retrieval & MRR
            ret_dirnc = '/home/franzihe/Documents/Data/Retrieval/%s/minute_surface_acc/' %station
            ret_f[int(year+month+day)] = netCDF4.Dataset(ret_dirnc+'%s%s%s.nc' %(year, month, day), 'r')

            ret_double_fence_p[int(year+month+day)] = ret_f[int(year+month+day)].variables['pulsed_sfc_acc'][:]
            ret_double_fence_u[int(year+month+day)] = ret_f[int(year+month+day)].variables['upslope_sfc_acc'][:]

            SWC_dirnc = '/home/franzihe/Documents/Data/Retrieval/%s/SWC_hourly/' %station
            SWC_f[int(year+month+day)] = netCDF4.Dataset(SWC_dirnc+'%s%s%s.nc' %(year, month, day), 'r')

            SWC_p[int(year+month+day)] = SWC_f[int(year+month+day)].variables['pulsed_SWC'][:]
            SWC_u[int(year+month+day)] = SWC_f[int(year+month+day)].variables['upslope_SWC'][:]

            MRR_dirnc = '/home/franzihe/Documents/Data/MRR/%s/' %station
 #       try:
            MRR_f[int(year+month+day)] = netCDF4.Dataset(MRR_dirnc+'MRR_HKL_%s%s%s_maahn_r01.nc' %(year,month,day), 'r')
            ZE[int(year+month+day)] = MRR_f[int(year+month+day)].variables['Ze'][:]
            MRR_height[int(year+month+day)] = MRR_f[int(year+month+day)]['height'][:]
            MRR_time[int(year+month+day)] = MRR_f[int(year+month+day)]['time'][:]
        except OSError:
            print('no data', int(year+month+day))
            continue
        

     #   ret_f[int(year+month+day)].close()
```

```python
def get_netCDF_variable(f, var_name, var, dim):
    v_0m = f.createVariable(varname=var_name, datatype=var.dtype, dimensions=dim, zlib=True)
    v_0m[:] = var[:]
    return(v_0m)
```

```python
for month in m:
    if month == '11':
        t = np.arange(9,31)
    if month == '12' or month == '01' or month == '03':
        t = np.arange(1,32)
    if month == '02':
        t = np.arange(1,29)
    if month == '11' or month == '12':
        year = '2016'
    if month == '01' or month == '02' or month == '03':
        year = '2017'
    

    for day in t:
        
        if day < 10:
            day = '0%s' %day
        else:
            day = '%s' %day
            
        # Connect today with the next day to colocate the observations and the MEPS forecast valid from 12UTC.
        if month == '11' and day == '30':
            year_next = year
            month_next = '12'
            day_next = '01'
        elif month == '12' and day == '31':
            year_next = '2017'
            month_next = '01'
            day_next = '01'
        elif month == '01'  and day == '31':
            month_next = '02'
            day_next = '01'
        elif month == '02' and day == '28':
            month_next = '03'
            day_next = '01'
        elif month == '03' and day >= '01':
            continue

        elif (int(day) + 1) < 10:
            day_next = '0%s' %(int(day) + 1)
            year_next = year
            month_next = month

        else:
            day_next = '%s' %(int(day) + 1)
            year_next = year
            month_next = month

        
        #Minutely observations
        _obs_12utc = int(obs_f[int(year+month+day)].variables['sec'].shape[0]/2)
        obs_dofe_acc[int(year+month+day)] = (np.concatenate((obs_double_fence_acc[int(year+month+day)][_obs_12utc:],
                                                             obs_double_fence_acc[int(year_next+month_next+day_next)][:_obs_12utc])))


        obs_wind_dir[int(year+month+day)] = (np.concatenate((obs_wind_dir_10m[int(year+month+day)][_obs_12utc:],
                                                             obs_wind_dir_10m[int(year_next+month_next+day_next)][:_obs_12utc])))

        obs_wind_speed[int(year+month+day)] = (np.concatenate((obs_wind_speed_10m[int(year+month+day)][_obs_12utc:],
                                                               obs_wind_speed_10m[int(year_next+month_next+day_next)][:_obs_12utc])))

        obs_air_temp[int(year+month+day)] = (np.concatenate((obs_double_fence_temp[int(year+month+day)][_obs_12utc:],
                                                             obs_double_fence_temp[int(year_next+month_next+day_next)][:_obs_12utc])))

        obs_time_[int(year+month+day)] = (np.concatenate((obs_time[int(year+month+day)][_obs_12utc:],
                                                          obs_time[int(year_next+month_next+day_next)][:_obs_12utc])))

        obs_f[int(year+month+day)].close()

        #Eklima
        _eklima_12utc = int(eklima_f[int(year+month+day)]['time_stamp'][:].shape[0]/2)
        eklima_dofe_acc[int(year+month+day)] = (np.concatenate((eklima_double_fence_acc[int(year+month+day)][_eklima_12utc:],
                                                                eklima_double_fence_acc[int(year_next+month_next+day_next)][:_eklima_12utc])))

        eklima_wind_dir[int(year+month+day)] = (np.concatenate((eklima_wind_dir_10m[int(year+month+day)][_eklima_12utc:],
                                                                eklima_wind_dir_10m[int(year_next+month_next+day_next)][:_eklima_12utc])))

        eklima_wind_speed[int(year+month+day)] = (np.concatenate((eklima_wind_speed_10m[int(year+month+day)][_eklima_12utc:],
                                                                  eklima_wind_speed_10m[int(year_next+month_next+day_next)][:_eklima_12utc])))

        eklima_air_temp[int(year+month+day)] = (np.concatenate((eklima_air_temp[int(year+month+day)][_eklima_12utc:],
                                                                eklima_air_temp[int(year_next+month_next+day_next)][:_eklima_12utc])))
        eklima_time_[int(year+month+day)] = (np.concatenate((eklima_time[int(year+month+day)][_eklima_12utc:],
                                                             eklima_time[int(year_next+month_next+day_next)][:_eklima_12utc])))

        eklima_f[int(year+month+day)].close()


         #Retrieval & MRR
        _ret_12utc_p = int(ret_f[int(year+month+day)].variables['pulsed_sfc_acc'].shape[0]/2)
        _ret_12utc_u = int(ret_f[int(year+month+day)].variables['upslope_sfc_acc'][:].shape[0]/2)


        ret_dofe_acc_p[int(year+month+day)] = (np.concatenate((ret_double_fence_p[int(year+month+day)][_ret_12utc_p:],
                                                               ret_double_fence_p[int(year_next+month_next+day_next)][:_ret_12utc_p])))

        ret_dofe_acc_u[int(year+month+day)] = (np.concatenate((ret_double_fence_u[int(year+month+day)][_ret_12utc_p:],
                                                               ret_double_fence_u[int(year_next+month_next+day_next)][:_ret_12utc_u])))

        ret_f[int(year+month+day)].close()

        _SWC_12utc_p = int(SWC_f[int(year+month+day)].variables['pulsed_SWC'].shape[1]/2)
        _SWC_12utc_u = int(SWC_f[int(year+month+day)].variables['upslope_SWC'].shape[1]/2)

        SWC_vert_p[int(year+month+day)] = (np.concatenate((SWC_p[int(year+month+day)][:,_SWC_12utc_p:],
                                                           SWC_p[int(year_next+month_next+day_next)][:,:_SWC_12utc_p]), axis=1))

        SWC_vert_u[int(year+month+day)] = (np.concatenate((SWC_u[int(year+month+day)][:,_SWC_12utc_u:],
                                                           SWC_u[int(year_next+month_next+day_next)][:,:_SWC_12utc_u]), axis=1))
        SWC_f[int(year+month+day)].close()


        

#        except KeyError:
 #           print(int(year+month+day))

         #write netCDF files
        _obs_filename = 'Met_%s%s%s_%s_%s.nc' %(year,month,day,int(_obs_12utc/60),station)
        _eklima_filename = 'precip_wind_temp_%s%s%s_%s.nc' %( year,month,day,_eklima_12utc)
        _ret_filename = '%s%s%s_%s.nc' %(year,month,day, int(_ret_12utc_p/60))

        cF.createFolder(obs_dirnc+ '/CTRL_ICE-T/')
        obs_fw = netCDF4.Dataset(obs_dirnc+ '/CTRL_ICE-T/' + _obs_filename, 'w')
        #create dimensions
        obs_fw.createDimension('time', obs_time_[int(year+month+day)].shape[0])
        obs_fw.createDimension('accum', obs_dofe_acc[int(year+month+day)].shape[1])


        obs_time_w      = get_netCDF_variable(obs_fw, 'sec', obs_time_[int(year+month+day)] , ('time'))
        obs_precip_w    = get_netCDF_variable(obs_fw, 'double_fence_accum', obs_dofe_acc[int(year+month+day)], ('time', 'accum'))
        obs_winddir_w   = get_netCDF_variable(obs_fw, '10m_wind_dir_mast1', obs_wind_dir[int(year+month+day)], ('time', ))
        obs_windspeed_w = get_netCDF_variable(obs_fw, '10m_wind_speed_mast1', obs_wind_speed[int(year+month+day)], ('time', ))
        obs_air_temp_w  = get_netCDF_variable(obs_fw, 'air_temp', obs_air_temp[int(year+month+day)], ('time', ))

        obs_fw.close()

        cF.createFolder(eklima_dirnc+ '/CTRL_ICE-T/')
        eklima_fw = netCDF4.Dataset(eklima_dirnc+ '/CTRL_ICE-T/' + _eklima_filename, 'w')
        #create dimensions
        eklima_fw.createDimension('time', eklima_time_[int(year+month+day)].shape[0])
        eklima_fw.createDimension('YYYYMMDD_HH', eklima_time_[int(year+month+day)].shape[1])

        eklima_time_w      = get_netCDF_variable(eklima_fw, 'time_stamp', eklima_time_[int(year+month+day)],('time', 'YYYYMMDD_HH' ))
        eklima_precip_w    = get_netCDF_variable(eklima_fw, 'RR_1', eklima_dofe_acc[int(year+month+day)], ('time', ))
        eklima_winddir_w   = get_netCDF_variable(eklima_fw, 'DD', eklima_wind_dir[int(year+month+day)], ('time', ))
        eklima_windspeed_w = get_netCDF_variable(eklima_fw, 'FF', eklima_wind_speed[int(year+month+day)], ('time', ))
        eklima_air_temp_w  = get_netCDF_variable(eklima_fw, 'TA', eklima_air_temp[int(year+month+day)], ('time', ))

        eklima_fw.close()

        try:
            _MRR_12utc = int(MRR_f[int(year+month+day)].variables['time'].shape[0]/2)
            Ze_vert[int(year+month+day)]  = (np.concatenate((ZE[int(year+month+day)][_MRR_12utc:,:],
                                                             ZE[int(year_next+month_next+day_next)][:_MRR_12utc,:]), axis =0))
            MRR_h[int(year+month+day)] = (np.concatenate((MRR_height[int(year+month+day)][_MRR_12utc:,:],
                                                          MRR_height[int(year+month+day)][:_MRR_12utc,:]), axis = 0))
            MRR_f[int(year+month+day)].close() 
        
        
            if MRR_h[int(year+month+day)].shape[0] >= ret_dofe_acc_p[int(year+month+day)].shape[0]:
                    cF.createFolder(ret_dirnc+ '/CTRL_ICE-T/')
                    ret_fw = netCDF4.Dataset(ret_dirnc+ '/CTRL_ICE-T/' + _ret_filename, 'w')
                    #create dimensions
               #     ret_fw.createDimension('time', ret_dofe_acc_p[int(year+month+day)].shape[0])
                    ret_fw.createDimension('time', ret_dofe_acc_p[int(year+month+day)].shape[0])
                    ret_fw.createDimension('height_retrieval', MRR_height[int(year+month+day)][:,3::2].shape[1])
                    ret_fw.createDimension('height_reflectivity', MRR_height[int(year+month+day)].shape[1] )


                    ret_precip_p_w  = get_netCDF_variable(ret_fw, 'pulsed_sfc_acc', ret_dofe_acc_p[int(year+month+day)], ('time',))
                    ret_precip_u_w  = get_netCDF_variable(ret_fw, 'upslope_sfc_acc', ret_dofe_acc_u[int(year+month+day)], ('time',))
                    swc_p_w         = get_netCDF_variable(ret_fw, 'pulsed_SWC', np.transpose(SWC_vert_p[int(year+month+day)]), ('time','height_retrieval'))
                    swc_u_w         = get_netCDF_variable(ret_fw, 'upslope_SWC', np.transpose(SWC_vert_u[int(year+month+day)]), ('time','height_retrieval'))
                    swc_height_w    = get_netCDF_variable(ret_fw, 'height_SWC', MRR_height[int(year+month+day)][:ret_dofe_acc_p[int(year+month+day)].shape[0],3::2], ('time', 'height_retrieval' ))
                    ze_w            = get_netCDF_variable(ret_fw, 'Ze', Ze_vert[int(year+month+day)][:ret_dofe_acc_p[int(year+month+day)].shape[0],:], ('time', 'height_reflectivity'))
                    ze_height_w     = get_netCDF_variable(ret_fw, 'height_Ze', MRR_height[int(year+month+day)][:ret_dofe_acc_p[int(year+month+day)].shape[0],:], ('time', 'height_reflectivity'))

                    ret_fw.close()
        except KeyError:
            print('no data', [int(year+month+day)])
            continue
#        except KeyError:
 #           print(int(year+month+day))
```
